## Importing Liabraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.feature_selection import SelectKBest , chi2
```

## Loading Data

```python
data = pd.read_csv('../global_heat_index.csv')
```

```python
data.head(20)
```

<font color = 'red'> here we can see that before 7 solar radiation is 0 ( as it is night) and after 18 (6pm) again SR is 0 as the sun sets<br><b> so this data is not at all needed so we will clean it


```python
data.info()
```

## <font color ='blue'> Cleaning Data

cleaning before 6 AM and after 6 PM data

```python
clean1 = data[data['Hour'] < 6]
clean2 = data[data['Hour'] > 18]
```

```python
#now we can drop this both from the main 'data'

data = data.drop(clean1.index , axis = 0)
data = data.drop(clean2.index , axis = 0)
```

```python
data.head(20)
```

<font color = 'red'> kept 6 and 7 hours so that if there is radiation in case we wont miss out

## Input and Output

```python
x = data.drop(['Solar Radiation (GHI)'], axis = 1)
y = data.iloc[:,10].values
 
col = x.keys()
col.tolist()
col
```

```python
#now we can also scale our data
```

```python
from sklearn.preprocessing import MinMaxScaler
x = pd.DataFrame(MinMaxScaler().fit(x).transform(x))

#already created col
```

# <font color ='green' > SelectKBest (chi square value)

<ul>
    <li> Using Chi Square Test</li>
    <li> Extracting top 4 features

```python
best_feature = SelectKBest(score_func= chi2, k = 10)
fit = best_feature.fit(x,y)

scores = pd.DataFrame(fit.scores_)
column = pd.DataFrame(col)

#combining both in the dataframe

feature_score = pd.concat([column,scores],axis=1)
feature_score.columns = ['Features','Score']
```

```python
feature_score.nlargest(10,'Score')
```

# Visualizing

```python
feature_score.nlargest(10,'Score').plot(kind = 'bar')
```
